```python
df1= df.sort_values(' Total Price ', ascending=False)
df1
```

```python
df.columns
```

```python
df2= df.groupby('gender')[' Total Price '].sum()
df2 
```

```python
color=['Blue', 'Black']
X_1=plt.bar(x=df2.index,height=df2.values, width=0.5, color= color)
plt.xlabel('gender')
plt.ylabel(' Total Price ')
plt.title ('Total spending by gender')
plt.ylim([0,160000000])
plt.bar_label(X_1,labels=df2.values,label_type='edge',padding=3)
```

```python
df3= df.groupby('category')[' Total Price '].sum()
df3.sort_values(ascending= False, inplace= True)
```

```python
df3
```

```python
df3.values
df31= (df3.values/df3.values.sum())*100
df31= df31.round(1)
df31


```

```python

color= ['Green','Blue','Red','Black','Yellow','Purple','Brown','Aqua']
X_2= plt.bar(x=df3.index,height=df31,color=color,width=0.6)
plt.bar_label(X_2,labels=df31,label_type='edge',padding=3)
plt.ylim([0,50])
plt.xticks(rotation='vertical')
plt.xlabel('category')
plt.ylabel('Percentage of shopping in each category')
plt.title ('Total spending by category in %')

```

```python
df9= df.groupby('gender')['category'].count()
df9
```

```python
color=['Aqua', 'Teal']
X_3=plt.bar(x=df9.index, height=df9.values, width=0.5, color= color)
plt.bar_label(X_3,labels=df9.values,label_type='edge',padding=3)
plt.ylim([0,70000])
plt.xlabel('gender')
plt.ylabel(' Purchase Count ')
plt.title ('Purchase Count by gender')
```

```python
df4= df.groupby('payment_method')[' Total Price '].sum()
df4
```

```python
df4.plot (kind = 'pie', autopct='%1.0f%%')

plt.title ('Payment method')
```

```python
df5= df.groupby('payment_method')['invoice_no'].count()
df5
```

```python
color=['Purple','Orange','lime']
X_4=plt.bar(x=df5.index, height=df5.values, width=0.5, color=color)
plt.bar_label(X_4,labels=df5.values,label_type='edge',padding=3)
plt.ylim([0,50000])
plt.xlabel('payment_method')
plt.ylabel(' Count of invoice ')
plt.title ('Count of invoice by each payment method')
```

```python
df6= df.groupby('shopping_mall')['invoice_no'].count()
df6
```

```python
df6.plot (kind = 'pie', autopct='%1.0f%%')
plt.title ('Favorite Shopping mall')
```

```python
df7= df.groupby('shopping_mall')[' Total Price '].sum()
df7.sort_values(ascending=True, inplace=True)
df7
```

```python
df71=df7.values/1000000
df71=df71.round(1)
df71
```

```python
color=['Green','Yellow','Lime','Orange','Black','Blue','Red','Purple','Aqua','Grey']
Y_5=plt.barh(y=df7.index, width=df71, color=color, height=0.8)
plt.bar_label(Y_5,labels=df71,label_type='edge',padding=3)
plt.xlim([0,60])
plt.ylabel('shopping_mall')
plt.xlabel(' Net purchase in millions ')
plt.title ('Net purchase in millions on each shopping mall ')
```
